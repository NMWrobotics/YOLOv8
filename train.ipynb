```python
import os
from ultralytics import YOLO
```

```python
# Load a pre-trained YOLOv8 model (you can change the model type if needed)
model = YOLO("yolov8s.pt")

# Train the model
results = model.train(data='/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/google_colab_config.yaml',
                      epochs=10,
                      imgsz=640,
                      batch=8,
                      project='/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/training_results',
                      name='football')
```

```python
import cv2


VIDEOS_DIR = os.path.join('/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/', 'videos')

video_path = os.path.join(VIDEOS_DIR, 'football_playing.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

# Load a model
model = YOLO('/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/training_results/football/weights/best.pt')  # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
