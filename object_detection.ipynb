```python
import os
from ultralytics import YOLO
import cv2
```

```python
VIDEOS_DIR = os.path.join('/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/', 'videos') #video path

video_path = os.path.join(VIDEOS_DIR, 'football_playing_top_view.mp4') #input video
video_path_out = '{}_out.mp4'.format(video_path) #output video

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

# Load the yolo model
model = YOLO('/content/drive/MyDrive/Colab Notebooks/YOLO Object Detection/training_results/football/weights/best.pt')

threshold = 0.5 #confidence threshold

while ret:
    # Perform inference on the frame
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        # Draw bounding box and label
        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            label = f"{model.names[int(class_id)].upper()}: {score:.2f}"
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)
    # Write the frame to the output video
    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
